4. 5. Summarise the mean number of calories in each `Beverage_category`.

```python
# method 1
drinks.groupby("Beverage_category")["Calories"].mean()
```

6. **Harder** - Which `Beverage_prep` type contains more than the average calories of all drinks?
(Hint: to answer this, you’ll have to first figure out what the average calories across all drinks are, and then use that as a filter for the grouped Beverage_prep data).

```python
# calculate mean calories across all drinks
average_cals = drinks["Calories"].mean()
average_cals
```

```python
# create a DataFrame of Beverage prep types and their mean calories
beverage_prep_cals = drinks.groupby("Beverage_prep")["Calories"].mean().reset_index(name = "mean_calories").copy()
beverage_prep_cals
```

```python
# subset beverage prep types where mean calories is larger than avg calories of all drinks
beverage_prep_cals.loc[beverage_prep_cals["mean_calories"] > average_cals]
```

7. **Harder**  - Which is the best type of **coffee drink** to get if you’re worried about consuming too many calories? (Hint: you can use pd.unique() on a column to get all unique values/levels of that column! Also remember that you can use isin() to subset with a list)

```python
# get unique drinks categories
pd.unique(drinks["Beverage_category"])
```

```python
# get drinks that have either the words coffee or espresso in them
coffee_drinks = drinks.loc[drinks["Beverage_category"].isin(['Coffee', 
                                                             'Classic Espresso Drinks', 
                                                             'Signature Espresso Drinks',
                                                             'Frappuccino® Blended Coffee', 
                                                             'Frappuccino® Light Blended Coffee'])].copy()
coffee_drinks
```

```python
coffee_drinks.sort_values("Calories").head(1)
```

8. Filter the dataset `drinks` so we only take "Classic Espresso Drinks", and save this in a new data frame called `espresso_drinks`.

```python
# Filter out espresso drinks
espresso_drinks = drinks.loc[drinks["Beverage_category"] == "Classic Espresso Drinks"].copy()
espresso_drinks
```

9. Group your `espresso_drinks` data frame by the type of beverage prep, and then find out the mean calories for each drink group.

```python
espresso_drinks.groupby("Beverage_prep").mean()
```

10. Get the same grouped mean `Calories` values as above, but this time sorted in descending order. (Hint: so you're able to sort the resulting groupby by the 'Calories' column you will want to ensure you get back a dataframe and not a series, so don't specify it's only `Calories` that you want to summarise, just do a groupby on all the whole dataframe and then arrange)

```python
espresso_drinks.groupby("Beverage_prep").mean().sort_values("Calories", ascending = False)
```

11. Extension: plots!

```python
import seaborn as sns
```

```python
ordered_data = drinks.groupby("Beverage_category")["Calories"].mean().sort_values(ascending = False).index
ordered_data
```

```python
sns.barplot(x = "Calories", 
            y = "Beverage_category", 
            data = drinks, 
            ci = False, 
            estimator = np.mean,
            order = ordered_data, 
            orient = "h")
```
